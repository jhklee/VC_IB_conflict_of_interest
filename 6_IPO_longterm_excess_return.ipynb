### industry ret to dict

```python
# average weight
def weighted_average2(sic_c,data_col,weight_col,by_col):
  # set df by sic
  df =  crsp_1995_2022_sic_dict[sic_c].copy()
  df.RET = df.RET.replace('C', np.NaN).astype(float)

  df['_data_times_weight'] = df[data_col]*df[weight_col]
  df['_weight_where_notnull'] = df[weight_col]*pd.notnull(df[data_col])

  g = df.groupby(by_col)
  result = g['_data_times_weight'].sum() / g['_weight_where_notnull'].sum()

  return result

  del df['_data_times_weight'], df['_weight_where_notnull'], df
```

```python
# industry returns by sic3
ind_ret_dict = dict()
for sic in tqdm(CRSP_sic3_list):
  t_df = pd.DataFrame(weighted_average2(sic,'RET','mkt_cap_million', 'date')).rename(columns={0:'ind_ret'}).reset_index()
  t_df.date =  pd.to_datetime(t_df.date)
  ind_ret_dict[sic] = t_df
```

```python
def long_ind_ret_func_dict(permno, idate,f_m_date, t_crsp, days):
  tempt_s1 = np.where(t_crsp.date == f_m_date)[0]
  tempt_s2 = np.where(t_crsp.date == idate + timedelta(days=days))[0]

  if len(tempt_s2)!=0:
    globals()['ind_long_ret_{}_dict'.format(days)][permno] = (1 + t_crsp.iloc[tempt_s1[0]:tempt_s2[0]+1].ind_ret).cumprod().iloc[-1] -1
  else:
    globals()['ind_long_ret_{}_dict'.format(days)][permno] = None
```
